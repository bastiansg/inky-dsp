```python
from time import sleep

from glob import glob
from PIL import Image
from rich.pretty import pprint

from inky_dsp.pipelines import display_image
from inky_dsp.buttons import InkyButtons, ButtonMap, ButtonEvent
```

```python
image_paths = glob("/resources/test-images/*.jpg")
print(f"image_paths: {len(image_paths)}")

image_paths_iter = iter(image_paths)
```

```python
image_path = next(image_paths_iter)
image = Image.open(image_path)

display_image(image=image.resize())
```

```python
async def on_button_pressed(event: ButtonEvent) -> None:
    pprint(event)
    sleep(5)
```

```python
buttons = InkyButtons(
    button_map=ButtonMap(),
    callback=on_button_pressed,
)
```

```python
await buttons.run()
```
